```python
import torch as th
import numpy as np
import os
from sacFFR import SACFFR_DIR, DATA_DIR
import sys
sys.path.append(os.path.join(SACFFR_DIR, "Train", "sac_modules")) 
def numpy2str(x: np.ndarray):
    if x.ndim==1:
        z =  ", ".join([str(t) for t in x])
    elif x.ndim==2:
        z = "\t\t"
        for i in range(x.shape[0]):
            z += ", ".join([str(t) for t in x[i]])
            if i < x.shape[0]-1:
                z += ",\n\t\t"
    else:
        raise Exception("Only vector and matrix are supported!!!")
    
    return z
net = th.load(os.path.join(DATA_DIR,"trained_sac", "norm_Actormodel_11_8_epi3550"))
net
```

```python
print(numpy2str(net.fc1.weight.detach().numpy()))
```

```python
print(numpy2str(net.fc1.bias.detach().numpy()))
```

```python
print(numpy2str(net.fc2.weight.detach().numpy()))
```
